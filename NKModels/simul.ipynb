```python
import econpizza as ep
import matplotlib.pyplot as plt
```

```python
model_opt = ep.load('optimal.yaml', raise_errors=True, verbose=True)
model_taylor = ep.load('taylor_.yaml', raise_errors=True, verbose=True)
```

## Modelo Ã³ptimo

```python
_ = model_taylor.solve_stst()
```

```python
model_opt.steady_state
```

```python
shk = ('emiu', 0.01)

x, flag = model_taylor.find_path(shock=shk)
```

```python
fig, axs = plt.subplots(2, 3, figsize=(10, 4))
for i,v in enumerate(("g", "i", "miu", "pi", "x")):

    axs.flatten()[i].plot(x[:35,model_opt['variables'].index(v)])
    axs.flatten()[i].set_xlabel(v)

fig.tight_layout()
```

```python
shk = ('emiu', 0.01 ** 2)

x, flag = model_opt.find_path(shock=shk)
```

```python
fig, axs = plt.subplots(2, 3, figsize=(10, 4))
for i,v in enumerate(("g", "i", "miu", "pi", "x")):

    axs.flatten()[i].plot(x[:35,model_opt['variables'].index(v)])
    axs.flatten()[i].set_xlabel(v)

fig.tight_layout()
```
